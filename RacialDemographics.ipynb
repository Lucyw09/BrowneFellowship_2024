```python
import pandas as pd
```

```python
pop_df = pd.read_csv("pgh_pop.csv")
pop_df.head()
```

```python
pop_df['2010_White_Alone_Population']/pop_df['2010_Total_Population']*100
```

```python
gar_df = pop_df.loc[[33]]
```

```python
poc_10 = 100 - gar_df['2010_White_Alone_Population']/gar_df['2010_Total_Population']*100
tot_10 = gar_df['2010_Total_Population']
# print(round(poc_10.iloc[0], 2))

poc_20 = 100 - gar_df['2020_White_Alone_Population']/gar_df['2020_Total_Population']*100
tot_20 = gar_df['2020_Total_Population']

print(f"in 2010, {round(poc_10.iloc[0], 2)}% of the total population in Garfield of {round(tot_10.iloc[0], 2)} identified as a person of color")
print(f"in 2020, {round(poc_20.iloc[0], 2)}% of the total population in Garfield of {round(tot_20.iloc[0], 2)} identified as a person of color")
```

```python
pop_tot_10 = pop_df["2010_Total_Population"].sum()
black1 = pop_df["2010_Black_Alone_Population"].sum()
native1 = pop_df["2010_American_Indian_and_Alaska_Native_alone_population"].sum()
asian1 = pop_df["2010_Asian_Alone_Population"].sum()
aapi1 = pop_df["2010_Native_Hawaiian_and_Other_Pacific_Islander_alone_Populatio"].sum()
other1 = pop_df["2010_Any_other_single_race_alone_Popualtion"].sum()
multi1 = pop_df["2010_Population_two_or_more_races"].sum()

pop_poc_10 = black1 + native1 + asian1 + aapi1 + other1 + multi1
print(f"In 2010, the total population in Pittsburgh was: {pop_tot_10}")
print(f"Of that total, {round(pop_poc_10/pop_tot_10, 4)*100}% identified as people of color")

pop_tot_20 = pop_df["2020_Total_Population"].sum()
black2 = pop_df["2020_Black_Alone_Population"].sum()
native2 = pop_df["2020_American_Indian_and_Alaska_Native_alone_population"].sum()
asian2 = pop_df["2020_Asian_Alone_Population"].sum()
aapi2 = pop_df["2020_Native_Hawaiian_and_Other_Pacific_Islander_alone_Populatio"].sum()
other2 = pop_df["2020_Any_other_single_race_alone_Popualtion"].sum()
multi2 = pop_df["2020_Population_two_or_more_races"].sum()

pop_poc_20 = black2 + native2 + asian2 + aapi2 + other2 + multi2

print(f"In 2020, the total population in Pittsburgh was: {pop_tot_20}")
print(f"Of that total, {round(pop_poc_20/pop_tot_20, 4)*100}% identified as people of color")
```

```python
import matplotlib.pyplot as plt

# df = gar_df.drop(['_id','Neighborhood', 'Change_2010_to_2020_Total_Population':'Change_2010_to_2020_Hispanic_or_Latino_Population'], axis=1)
# df = gar_df.loc[:, "2010_White_Alone_Population", "2010_Black_Alone_Population", "2010_American_Indian_and_Alaska_Native_alone_population", "2010_Asian_Alone_Population", "2010_Native_Hawaiian_and_Other_Pacific_Islander_alone_Populatio", "2010_Any_other_single_race_alone_Popualtion", "2010_Population_two_or_more_races", "2010_Hispanic_or_Latino_Population"]
# print(df)
df_2010 = pop_df.iloc[:, 4:11]
# 14:21 is 2020
df_2010_sum = df_2010.sum()
# print(df_2010_sum)
# pop_df_sum.dtypes

# df.head()
print(df_2010.ndim)  #prints 2D, therefore needs to be converted to 1D to use pie chart
# print(df.iloc[0].ndim)
# print(df.iloc[0])
label = ["White", "Black", "Native", "Asian", "AAPI", "Other", "Two or more"]
```

```python
import numpy as np
x = np.char.array(["White", "Black", "Native", "Asian", "AAPI", "Other", "Two or more"])
y = df_2010_sum.values.flatten()
colors = ['violet','lightskyblue','cyan','teal', 'darkgreen','gold','lightgreen']

# plt.pie(df_2010_sum.values.flatten(), labels = label)
porcent = 100.*y/y.sum()

patches, texts = plt.pie(y, colors=colors, startangle=90, radius=1.2)
labels = ['{0} - {1:1.2f} %'.format(i,j) for i,j in zip(x, porcent)]

sort_legend = True
if sort_legend:
    patches, labels, dummy =  zip(*sorted(zip(patches, labels, y),
                                          key=lambda x: x[2],
                                          reverse=True))

plt.legend(patches, labels, loc='best', bbox_to_anchor=(-0.1, 1.),
           fontsize=8)
plt.title("Pittsburgh Demographics in 2010")

# plt.savefig('piechart.png', bbox_inches='tight')
```
